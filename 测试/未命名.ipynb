```python
'''
配货补货差：0-5,5-10，10-20,20-30,30-50,50以上
时间差：0-10,10-20,20-30,30-50,50-80,80以上
补货周期：
点位日均销售额：0-200,200-500,500-1000,1000-1500,1500-2000,2000以上
'''

D=[]

for i in df['补货配货差'].tolist():
    if i< 5:
        score='0-5'
    elif 5<=i<10:
        score='5-10'
    elif 10<=i<20:
        score='10-20'
    elif 20<=i<30:
        score='20-30'
    elif 30<=i<50:
        score='30-50'
    else:
        score='50以上'
    D.append(score)
df['补货配货差范围']=D
df
```

```python
'''
配货补货差：0-5,5-10，10-20,20-30,30-50,50以上
时间差：0-10,10-20,20-30,30-50,50-80,80以上
补货周期：
点位日均销售额：0-200,200-500,500-1000,1000-1500,1500-2000,2000以上
'''

D=[]

for i in df['时间差'].tolist():
    if i< 10:
        score='0-10'
    elif 10<=i<20:
        score='10-20'
    elif 20<=i<30:
        score='20-30'
    elif 30<=i<50:
        score='30-50'
    elif 50<=i<80:
        score='50-80'
    else:
        score='80以上'
    D.append(score)
df['时间差范围']=D
df
```
